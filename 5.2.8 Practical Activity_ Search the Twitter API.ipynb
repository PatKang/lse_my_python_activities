### Cross-check trending topics between NY and London

* If bitcoin is not one of the common topics what are the common trending topics between the two cities?

```python
# Prepare the workstation by importing the libraries

import yaml 
from yaml.loader import SafeLoader

twitter_creds = yaml.safe_load(open('twitter_tmp.yaml', 'r').read())

twitter_creds.keys()

from twitter import *

# Now we can define the access token, access token secret, api key and api key secret. Keep the Order as that.
twitter_api = Twitter(auth = OAuth(twitter_creds['access_token'],
                                   twitter_creds['access_token_secret'],
                                   twitter_creds['api_key'],
                                   twitter_creds['api_secret_key']))

# Check if we have an active connection
print(twitter_api)
```

```python
# Let's test the connection by checking the #python in twitter
python_tweets = twitter_api.search.tweets(q = '#python')

# View the tweets
print(python_tweets)
```

***Since we have an active connection lets now get to work***

# Identify Trends worldwide

```python
# Identify a worldwide trend
trends_worldwide = twitter_api.trends.available()

# How many trends are available?
print(len(trends_worldwide))

# Example of trends_worldwide
trends_worldwide[1]
```

```python
# Now we should find the woeid of NY 
nyc = 'New York'

# Create a variable to store the list of names
list_of_names_ny = [_ for _ in trends_worldwide if _['name'] == nyc]

# View the output
print(len(list_of_names_ny))

# Use the index to find NY
list_of_names_ny[0]
```

```python
# Now we find the ID for London
london = 'London'

# Create a variable to store the list of names
list_of_names_london = [_ for _ in trends_worldwide if _['name'] == london]

# View the output
print(len(list_of_names_london))

# use index to find london
list_of_names_london[0]
```

### Find list of trends specifiy to both cities and save it to a PD DF

* Limit the output to more than 50k tweets

```python
# Let's start with NY, create a variable for that
ny_trends = twitter_api.trends.place(_id = list_of_names_ny[0]['woeid'])

# View the output
print(ny_trends)
```

```python
# Let's save this in a pandas dataframe 
import pandas as pd

# Create a DataFrame
ny_trends_df = pd.DataFrame(ny_trends[0]['trends'])

# Limit the output to 50k tweets
ny_trends_50k = ny_trends_df[ny_trends_df['tweet_volume'] > 50000].sort_values('tweet_volume', ascending = False)

print(ny_trends_50k)

# Save the list of popular tweets as a csv 
ny_trends_50k.to_csv('ny_trends_over50k.csv', index = False)
```

```python
# Repeat with London, to keep it concise we will do all in one code cell

# Find trends and store it in a variable
london_trends = twitter_api.trends.place(_id = list_of_names_london[0]['woeid'])

# Create a DataFrame
london_trends_df = pd.DataFrame(london_trends[0]['trends'])

# Limit the DataFrame to tweets over 50k 
london_trends_concise = london_trends_df[london_trends_df['tweet_volume'] > 50000]\
.sort_values('tweet_volume', ascending = False)

# Print the output
print(london_trends_concise)

# Save the DataFrame to csv
london_trends_concise.to_csv('london_tweets_over50k.csv', index = False)
```

# Find common trends between NY and London

```python
# Get the WOEID for the cities
print(list_of_names_ny[0]['woeid'])
print(list_of_names_london[0]['woeid'])
```

```python
# Look at trends in New York.
new_york_trends = twitter_api.trends.place(_id = list_of_names_ny[0]['woeid'])

# View the output.
new_york_trends
```

```python
# Create a variable to store the trends
ny_trend_list = [trend['name'] for trend in new_york_trends[0]['trends']]

# View the output
print(ny_trend_list)
```

```python
# Repeat for london
london_trends = twitter_api.trends.place(_id = list_of_names_london[0]['woeid'])

# Create a variable to store the trends
london_trend_list = [trend['name'] for trend in london_trends[0]['trends']]

# View the output
print(london_trend_list)
```

```python
# Find common trends between the two cities

# Create sets for the trend lists 
ny_trend_set = set(ny_trend_list)
london_trend_set = set(london_trend_list)

# Set a variable that compares both sets
common_trends = ny_trend_set.intersection(london_trend_set)

# View the output
print(common_trends)
```
