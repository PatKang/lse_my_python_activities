# Time Series Forecasting

* We will plot the sub-data sets
* Resample to remove noise
* Use a time-series moving average

```python
# Create a plot for 1 bedroom
# Calculate the max and min price
data_oneroom_non_outlier['price'].min()
data_oneroom_non_outlier['price'].max()

# Plot the time series data
data_oneroom_non_outlier.plot(figsize = (12, 4))
plt.legend(loc = 'best')
plt.title('Time Series plot for house with 1 bedroom')
plt.show(block = False)
```

```python
# Resample the dataset with one bedrrom and remove the missing values
one_room_resampled = data_oneroom_non_outlier.resample('M').mean()
one_room_resampled.head()

# Drop the missing values
one_room_resampled.dropna(inplace = True)
one_room_resampled.isna().sum()

# Plot the time-series data:
one_room_resampled.plot(figsize = (12, 4))
plt.legend(loc = 'best')
plt.title('Time series plot after resampling')
plt.show(block = False)
```

```python
# Calculate and plot the simple moving average

def plot_moving_average(series, window, plot_intervals = False, scale = 1.96):
    
    rolling_mean = series.rolling(window = window).mean()
    
    plt.figure(figsize = (12, 4))
    plt.title('Moving Average\n window size = {}'.format(window))
    plt.plot(rolling_mean, 'g', label = 'Simple moving average trend')
    
    # Plot confidence intervals for smoothed values
    if plot_intervals:
        mae = mean_absolute_error(series[window:], rolling_mean[window:])
        deviation = np.std(series[window:] - rolling_mean[window:])
        lower_bound = rolling_mean - (mae + scale * deviation)
        upper_bound = rolling_mean + (mae + scale * deviation)
        plt.plot(upper_bound, 'r--', label = 'Upper bound / Lower bound')
        plt.plot(lower_bound, 'r--')
        
    plt.plot(series[window:], label = 'Actual values')
    plt.legend(loc = 'best')
    plt.grid(True)
```

```python
# Plot the moving averages with five days smoothing
plot_moving_average(one_room_resampled.price, 5, plot_intervals = True)
```

```python
# Plot the moving averages with 30 days smoothing (monthly)
plot_moving_average(one_room_resampled.price, 30, plot_intervals = True)
```

```python
# Plot the moving averages with 90 days smoothing (quarterly)
plot_moving_average(one_room_resampled.price, 90, plot_intervals = True)
```
