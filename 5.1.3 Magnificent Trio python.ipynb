```python
# Set up the workstation 

import pandas as pd
import numpy as np
```

```python
# Create a dataframe from the csv file 

transport_cost = pd.read_excel('transport_costs.xlsx')

# Check the datatypes 
print(transport_cost.dtypes)
print(transport_cost.shape)
transport_cost
```

```python
transport_cost['total_sea_freight_costs'] = transport_cost['sea freight cost'] * 2 * 4

transport_cost
```

```python
transport_cost['road_transport_cost'] = transport_cost['road transport cost per km'] * 10000

transport_cost['total_export_costs'] = transport_cost['road_transport_cost'] + transport_cost['total_sea_freight_costs']

transport_cost
```

```python
transport_cost_export = transport_cost[['region', 'country', 'port', 'sea freight cost',
                                        'road transport cost per km', 'total_sea_freight_costs', 
                                        'road_transport_cost', 'total_export_costs']]

transport_cost_export.to_csv('5.1.3_python.csv', index = False)
```
